```python
# libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("Datasets/titanic.csv")
```

```python
titanic.drop("Name", inplace=True, axis=1)
titanic.drop("Ticket", inplace=True, axis=1)
```

```python
x = titanic.iloc[:, 2:]
y = titanic.iloc[:, 1]
```

```python
sex = pd.get_dummies(x['Sex'], drop_first = True)
x = x.drop('Sex', axis = 1)
x = pd.concat([x, sex], axis = 1)
```

```python
cabin = pd.get_dummies(x['Cabin'], drop_first = True)
x = x.drop('Cabin', axis = 1)
x = pd.concat([x, cabin], axis = 1)
```

```python
embark = pd.get_dummies(x['Embarked'], drop_first = True)
x = x.drop('Embarked', axis = 1)
x = pd.concat([x, embark], axis = 1)
```

```python
x = x.dropna(axis="columns", how="any")
x
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 3)
classifier.fit(x_train, y_train)
```

```python
y_pred = classifier.predict(x_test)
from sklearn.metrics import accuracy_score, classification_report
print("Accuracy =", accuracy_score(y_test, y_pred))
```
