```python
# importing statements

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# for 3D plots

from mpl_toolkits.mplot3d import Axes3D
from sklearn import datasets
from sklearn.decomposition import PCA
```

```python
iris = pd.read_csv("Datasets/iris.csv") # reading csv file

iris.info() # getting info about the iris dataset
```

```python
# statistical insight

iris.describe()
```

```python
# stripplot

sns.stripplot(x="species", y="petal_length", palette="husl", data=iris)
sns.stripplot(x="species", y="petal_length", data=iris, color="w", alpha=.9);
```

```python
# plot histogram

iris.hist(edgecolor='black', linewidth=1.2)
fig=plt.gcf()
fig.set_size_inches(12,6)
plt.show() # to plot the graph
```

```python
# draws heatmap with input as the correlation matrix calculted by(iris.corr())

plt.figure(figsize = (8, 6)) # size of map
sns.heatmap(iris.corr(), annot=True, cmap='cubehelix_r') # cmap = color set, annot = values written
plt.show()
```

```python
# create scatter plot

iris = sns.load_dataset('iris')

# style used as a theme of graph
# for example if we want black
# graph with grid then write "darkgrid"
sns.set_style("whitegrid")

# sepal_length, petal_length are iris
# feature data height used to define
# Height of graph whereas hue store the
# class of iris dataset.
sns.FacetGrid(iris, hue ="species", height = 6).map(plt.scatter, 'sepal_length', 'petal_length').add_legend()
```

```python
plt.plot(iris["sepal_length"], "r--")
plt.show()
```

```python
# 3D plot

X = iris.data[:, :2]  # we only take the first two features.
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(2, figsize=(8, 6))
plt.clf()

# Plot the training points
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Sepal length")
plt.ylabel("Sepal width")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

# To get a better understanding of interaction of the dimensions plot the first three PCA dimensions
fig = plt.figure(1, figsize=(8, 6))
ax = Axes3D(fig, elev=-150, azim=110)
X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)
ax.set_title("First three PCA directions")
ax.set_xlabel("1st eigenvector")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("2nd eigenvector")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("3rd eigenvector")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
